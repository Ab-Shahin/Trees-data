```python
# filling up all the na values of the stump and dead trees with 'Not Applicable'
mask = ((tree_census_st['status'] == 'Stump') | (tree_census_st['status'] == 'Dead'))
tree_census_st.loc[mask] = tree_census_st.loc[mask].fillna('Not Applicable')
```

```python
# Excluding all the extreme diameters values to calculate a more accurate values' means and maxes 
tree_census_st = tree_census_st[(tree_census_st['tree_dbh']<=60) & (tree_census_st['stump_diam']<=60)]
tree_census_st
```

```python
# filling up na values with most common or logical values
tree_census_st['problems'].fillna('None', inplace=True)
tree_census_st['health'].fillna('Good', inplace=True)
tree_census_st['spc_latin'].fillna('No Observation', inplace=True)
tree_census_st['sidewalk'].fillna('NoDamage', inplace=True)
tree_census_st['spc_common'].fillna('No Observation', inplace=True)
tree_census_st.isna().sum()
```

```python
# creating alive and dead data subsets
trees_stump_alive = tree_census_st[tree_census_st['status'] == 'Alive']
trees_stump_or_dead = tree_census_st[(tree_census_st['status'] == 'Stump') | (tree_census_st['status'] == 'Dead')]
```

```python
alive_stats = trees_stump_alive.groupby('spc_latin')['tree_dbh'].describe()
alive_stats
```

```python
# merging the alive dataset with stats subset to correct values deviations for tree_dbh
trees_stump_alive = trees_stump_alive.merge(alive_stats, on='spc_latin', how='left')
trees_stump_alive
```

```python
mask = trees_stump_alive['tree_dbh']< trees_stump_alive['25%']
trees_stump_alive.loc[mask,'tree_dbh' ] = trees_stump_alive['25%']

mask = trees_stump_alive['tree_dbh'] > trees_stump_alive['75%']
trees_stump_alive.loc[mask, 'tree_dbh'] = trees_stump_alive['75%']
```

```python
# tree_dbh values have been adjusted to fit within 25 and 75 percentiles 
trees_stump_alive
```
